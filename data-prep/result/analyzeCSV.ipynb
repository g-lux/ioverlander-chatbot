```python
from IPython.display import display, HTML
import pandas as pd

df = pd.read_csv('outputWithMapURL.csv')

# Function to display a row of the dataframe along with the image
def display_row_with_image(row):
    display(HTML(f"""
    <div style="margin-bottom: 20px;">
        <h4>{row['Name']}</h4>
        <p><strong>Location:</strong> {row['Location']}</p>
        <p><strong>Description:</strong> {row['Description']}</p>
        <p><strong>Coordinates:</strong> Latitude {row['Latitude']}, Longitude {row['Longitude']}</p>
        <p><strong>Altitude:</strong> {row['Altitude']} meters</p>
        <img src="{row['maps']}" style="max-width: 400px; max-height: 300px;">
    </div>
    """))

# Display the first 2000 rows along with their images
for index, row in df.head(2000).iterrows():
    display_row_with_image(row)
```
